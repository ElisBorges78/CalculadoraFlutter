```python
import random

# Conjunto de desenhos da forca
boneco_na_forca = [
    """
    ------
    |    |
    |
    |
    |
    |
    """,
    """
    ------
    |    |
    |    O
    |
    |
    |
    """,
    """
    ------
    |    |
    |    O
    |    |
    |
    |
    """,
    """
    ------
    |    |
    |    O
    |   /|
    |
    |
    """,
    """
    ------
    |    |
    |    O
    |   /|\\
    |
    |
    """,
    """
    ------
    |    |
    |    O
    |   /|\\
    |   /
    |
    """,
    """
    ------
    |    |
    |    O
    |   /|\\
    |   / \\
    |
    """,
]

def escolher_palavra():
    # Lista de palavras para escolher
    palavras = ["python", "programacao", "jogo", "computador", "forca", "desenvolvimento", "linguagem"]
    # Escolher uma palavra aleatória
    return random.choice(palavras)

def exibir_status(palavra, letras_adivinhadas):
    # Exibe o status da palavra, substituindo letras não adivinhadas por "_"
    status = ""
    for letra in palavra:
        if letra in letras_adivinhadas:
            status += letra + " "
        else:
            status += "_ "
    print(status)

def jogo_da_forca():
    palavra = escolher_palavra()  # Escolher uma palavra aleatória
    tentativas_restantes = 6  # Número de tentativas permitidas (6 desenhos de boneco)
    letras_adivinhadas = set()  # Conjunto para armazenar letras adivinhadas pelo jogador

    print("Bem-vindo ao Jogo da Forca!")
    print("A palavra tem", len(palavra), "letras.")

    while tentativas_restantes > 0:
        # Exibir estado atual da forca
        print(boneco_na_forca[6 - tentativas_restantes])

        # Exibir status da palavra
        exibir_status(palavra, letras_adivinhadas)

        # Solicitar uma letra ao jogador
        letra = input("Digite uma letra: ").lower()

        # Verificar se a letra já foi adivinhada
        if letra in letras_adivinhadas:
            print("Você já tentou essa letra. Tente outra.")
            continue

        # Adicionar a letra às tentativas feitas
        letras_adivinhadas.add(letra)

        # Verificar se a letra está na palavra
        if letra in palavra:
            print("Correto! A letra está na palavra.")
            # Verificar se o jogador adivinhou todas as letras
            if all(letra in letras_adivinhadas for letra in palavra):
                print("Parabéns! Você adivinhou a palavra:", palavra)
                break
        else:
            # Reduzir tentativas restantes se a letra não estiver na palavra
            tentativas_restantes -= 1
            print("Errado! Você perdeu uma tentativa.")
            print(f"Tentativas restantes: {tentativas_restantes}")

        # Verificar se as tentativas acabaram
        if tentativas_restantes == 0:
            print(boneco_na_forca[-1])
            print("Você perdeu! A palavra era:", palavra)

# Iniciar o jogo
jogo_da_forca()
```
